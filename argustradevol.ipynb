```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plotting(grade):
    df = pd.read_excel('tradevolprocessingplot.xlsx', sheet_name=grade)
    df.plot(drawstyle="steps-post", linewidth=2,title=grade,figsize=(10,6)  )
```

```python
plotting('CASH ROLL')
plotting('DSW CMA')
plotting('MDY')
plotting('MEH')
plotting('LLS')
plotting('MARS')
plotting('WCS-USGC')
plotting('WCS-CUSH')
```

```python
def plotting(grade):
    df = pd.read_excel('tradevolprocessingplot.xlsx', sheet_name=grade)
    fig, ax = plt.subplots(figsize=(10,8))
    def plothelper(mmm,farge):
        df.plot(x='index', y=mmm,color=farge, ax=ax,drawstyle="steps-post",linewidth=2)
    def plothelper2(mmm):
        df.plot(x='index', y=mmm,ax=ax,drawstyle="steps-post",linewidth=2)
        
    df.plot(x='index', y='Sep',color='#666666', ax=ax,drawstyle="steps-post",linewidth=5)
    plothelper('Aug','#C00000')
    plothelper('Jul','#003366')
    plothelper('Jun','#669933')
    plothelper2('May')
    plothelper2('Apr')

    plt.ylabel("kbd")
    plt.xlabel("days")
    plt.title(grade)
    plt.show()

plotting('CASH ROLL')
plotting('DSW CMA')
plotting('MDY')
plotting('MEH')
plotting('LLS')
plotting('MARS')
plotting('WCS-USGC')
plotting('WCS-CUSH')
```
